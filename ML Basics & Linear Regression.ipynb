Question 3

What is Linear Regression? Write the Linear Regression equation and explain intercept and coefficient.

Answer:

Linear Regression is a supervised machine learning algorithm used to predict a continuous numerical value by finding a linear relationship between the independent variables (features) and the dependent variable (target).

Linear Regression Equation:

y=b0​+b1​x1​+b2​x2​+⋯+bn​xn

Where:

y = Predicted value (Target)
b₀ = Intercept
b₁, b₂, …, bₙ = Coefficients
x₁, x₂, …, xₙ = Independent variables (Features)
Explanation
Intercept (b₀): The predicted value of the target when all independent variables are 0.
Coefficient (b₁, b₂, …): Shows how much the target value changes when a feature increases by one unit, while keeping other features constant.

Example:
If the coefficient of RM (number of rooms) is positive, then increasing the number of rooms generally increases the predicted house price.

Question 4

Explain the following evaluation metrics used in Linear Regression:

Answer:

1. SSE (Sum of Squared Errors)

SSE is the sum of the squared differences between the actual values and the predicted values.
A lower SSE indicates a better model.

2. MSE (Mean Squared Error)

MSE is the average of the squared errors.
Lower MSE means better prediction accuracy.

3. RMSE (Root Mean Squared Error)

RMSE is the square root of MSE.
It measures the average prediction error in the same unit as the target variable.
Lower RMSE indicates a better model.

4. R² Score (Coefficient of Determination)

R² measures how well the model explains the variation in the target variable.
Its value ranges from 0 to 1.
A value closer to 1 indicates a better model.

5. Adjusted R²

Adjusted R² is a modified version of R² that considers the number of independent variables.
It helps prevent overestimating model performance when more features are added.

Explain the Machine Learning workflow steps from data collection to model evaluation.

Answer:
Data Collection
Collect relevant data from sources such as databases, CSV files, APIs, or websites.

Data Preprocessing
Clean the data by handling missing values, removing duplicates, and correcting inconsistent data.
Convert data into the required format if necessary.

Exploratory Data Analysis (EDA)
Analyze the dataset using summary statistics and visualizations.

Feature Selection
Select the most important independent variables (features) that influence the target variable.

Data Splitting
Split the dataset into training and testing sets (commonly 80% training and 20% testing).

Model Training
Train the Machine Learning model using the training data.

Model Prediction

Tune the model or improve feature selection to achieve better performance if required.

```python
# Question 6

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/BostonHousing.csv')

print("First 5 Rows:")
print(df.head())

print("\nDataset Information:")
df.info()

print("\nSummary Statistics:")
print(df.describe())
```

Question 6 - Interpretation

The dataset was successfully loaded into Python.
The first five rows provide a preview of the data.
The dataset information shows the number of rows, columns, data types, and missing values.
The summary statistics provide the mean, minimum, maximum, and standard deviation of the numerical variables.

```python
# Question 7

X = df.drop("medv", axis=1)   # Independent variables
y = df["medv"]                # Dependent variable (House Price)

print("Independent Variables (X):")
print(X.head())

print("\nDependent Variable (y):")
print(y.head())
```

Question 7 - Interpretation

The dataset was divided into independent variables (X) and the dependent variable (y).
The independent variables contain the input features, while 'medv' is the target variable representing the median house value.

```python
# Question 8

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)

print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
```

Question 8 - Interpretation

The dataset was split into training and testing sets.
80% of the data was used to train the model, while the remaining 20% was reserved for testing the model's performance.
The random_state value ensures that the split remains the same each time the code is executed.

```python
# Question 9

from sklearn.linear_model import LinearRegression

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

print("First 10 Predicted Values:")
print(y_pred[:10])
```

Question 9 - Interpretation

A Linear Regression model was trained using the training dataset.
The trained model was then used to predict house prices for the testing dataset.
The predicted values represent the estimated house prices based on the input features.

```python
# Question 10

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R² Score:", r2)
```

Question 10 - Interpretation

The Linear Regression model was evaluated using MAE, MSE, RMSE, and R² Score.
Lower values of MAE, MSE, and RMSE indicate better prediction accuracy.
The R² Score shows how well the model explains the variation in house prices. A value closer to 1 indicates better model performance.
