```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/House Price India.csv')

df.info()
df.describe()
```

```
#Ques-1
# Question 1
average_price = df["Price"].mean()
median_price = df["Price"].median()
std_price = df["Price"].std()

print("Average Price:", average_price)
print("Median Price:", median_price)
print("Standard Deviation:", std_price)
```

Question 1 - Interpretation

• The average house price shows the overall mean price.
• The median represents the middle value and is less affected by outliers.
• The standard deviation indicates the variation in house prices. A higher value means prices vary more.

```
# Question 2

most_common_bedrooms = df["number of bedrooms"].mode()[0]
print("Most common number of bedrooms:", most_common_bedrooms)
```

Question 2 - Interpretation

The most common number of bedrooms is 3.
This indicates that most houses in the dataset have 3 bedrooms.

```
# Question 3

print("Skewness of Price:", df["Price"].skew())
```

Question 3 - Interpretation

If the skewness value is greater than 0, the house prices are positively (right) skewed.
If the skewness value is less than 0, the house prices are negatively (left) skewed.
If the skewness value is close to 0, the price distribution is approximately symmetric.

```
# Question 4

avg_price_bedrooms = df.groupby("number of bedrooms")["Price"].mean()
print(avg_price_bedrooms)
```

Question 4 - Interpretation

The average house price was calculated for each bedroom category.
The results show how the average price changes as the number of bedrooms increases.
This helps compare the average price of houses with different numbers of bedrooms.

```
# Question 5

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df["Area of the house(excluding basement)"], df["Price"])
plt.xlabel("Area of House (Excluding Basement)")
plt.ylabel("House Price")
plt.title("Relationship Between Living Area and House Price")
plt.show()
```

Question 5 - Interpretation

The scatter plot shows a positive relationship between the living area and house price.
In general, houses with larger living areas tend to have higher prices, although there are some variations.

```
# Question 6

anomalies = df[
    (df["Price"] > df["Price"].quantile(0.90)) &
    (df["Area of the house(excluding basement)"] < df["Area of the house(excluding basement)"].quantile(0.25))
]

print(anomalies[["Price", "Area of the house(excluding basement)", "number of bedrooms"]])
```

Question 6 - Interpretation

Some houses have unusually high prices despite having relatively small living areas.
These properties may be expensive because of factors such as location, luxury features, waterfront access, or other premium amenities.

```
# Question 7

avg_price = df.groupby(["number of floors", "waterfront present"])["Price"].mean()

print(avg_price)
```

Question 7 - Interpretation

The average house price was compared based on the number of floors and whether the house has waterfront access.
In general, houses with waterfront access have higher average prices than those without waterfront access.
The number of floors also influences the average house price.

```
# Question 8

min_price = df["Price"].min()
max_price = df["Price"].max()

print("Minimum House Price:", min_price)
print("Maximum House Price:", max_price)
```

Question 8 - Interpretation

The dataset contains both low-priced and high-priced houses.
The difference between the minimum and maximum prices indicates that house prices vary significantly depending on factors such as area, location, number of bedrooms, and amenities like waterfront access.

```
# Question 9

correlation = df["Price"].corr(df["Area of the house(excluding basement)"])

print("Correlation between House Price and Living Area:", correlation)
```

Question 9 - Interpretation

The correlation value measures the relationship between house price and living area.
A positive correlation indicates that houses with larger living areas generally have higher prices.
A stronger correlation means the relationship between area and price is stronger.

Question 10 – Final Insights

The average house price is around the overall market value, while the median helps reduce the effect of extreme prices.

The most common houses in the dataset have 3 bedrooms.

House prices generally increase as the living area (excluding basement) increases.

Houses with waterfront access tend to have higher average prices than houses without waterfront access.

There is a noticeable variation between the minimum and maximum house prices, showing that property prices differ significantly based on factors such as area, bedrooms, floors, location, and amenities.
