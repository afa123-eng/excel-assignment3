```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Titanic.csv')

df.info()
df.describe()
```

```python
# Question 1

plt.figure(figsize=(6,4))
sns.countplot(x="Sex", data=df)
plt.title("Number of Passengers by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.show()
```

Question 1 - Interpretation

The countplot shows the number of male and female passengers on the Titanic.
It helps compare the gender distribution of passengers in the dataset.

```python
# Question 2

plt.figure(figsize=(6,4))
sns.countplot(x="Pclass", data=df, palette="Set2")
plt.title("Number of Passengers by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.show()
```

Question 2 - Interpretation

The bar chart shows the number of passengers in each passenger class.
It helps compare the distribution of passengers across the three travel classes.

```python
# Question 3

plt.figure(figsize=(6,4))
sns.histplot(df["Age"], bins=20, kde=True, color="skyblue")
plt.title("Distribution of Passenger Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Question 3 - Interpretation

The histogram shows the distribution of passenger ages.
Most passengers were young and middle-aged, while very few passengers were at the extreme age groups.
The KDE curve helps visualize the overall age distribution.

```python
# Question 4

plt.figure(figsize=(6,5))
sns.boxplot(x="Survived", y="Age", data=df)
plt.title("Age Distribution by Survival Status")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```

Question 4 - Interpretation

The box plot compares the age distribution of passengers who survived and those who did not survive.
It helps identify differences in median age, spread, and possible outliers between the two groups.

```python
# Question 5

plt.figure(figsize=(6,5))
sns.scatterplot(x="Age", y="Fare", data=df)
plt.title("Relationship Between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

Question 5 - Interpretation

The scatter plot shows the relationship between passenger age and fare.
There is no strong linear relationship between age and fare, although some passengers paid much higher fares than others.
The chart also highlights a few outliers with very high fares.

```python
# Question 6

survival_counts = df["Survived"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    survival_counts,
    labels=["Did Not Survive", "Survived"],
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Survival Percentage of Passengers")
plt.show()
```

Question 6 - Interpretation

The pie chart shows the percentage of passengers who survived and those who did not survive.
It provides a clear comparison of the survival distribution in the Titanic dataset.

```python
# Question 7

plt.figure(figsize=(10,8))

numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```

Question 7 - Interpretation

The heatmap shows the correlation between numerical variables in the Titanic dataset.
Positive values indicate a positive relationship, while negative values indicate an inverse relationship.
Variables with correlation values close to 1 or -1 have a stronger relationship than those close to 0.

```python
# Question 8

selected_columns = ["Age", "Fare", "SibSp", "Parch"]

sns.pairplot(df[selected_columns].dropna())

plt.show()
```

Question 8 - Interpretation

The pair plot shows the relationship between Age, Fare, SibSp, and Parch.
It helps identify patterns, trends, and the distribution of each numerical variable.
The diagonal plots show the distribution of each variable, while the other plots show relationships between variables.

```python
# Question 9

plt.figure(figsize=(7,5))

sns.violinplot(x="Pclass", y="Fare", data=df)

plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

Question 9 - Interpretation

The violin plot compares the fare distribution across different passenger classes.
It shows that first-class passengers generally paid higher fares, while second- and third-class passengers paid lower fares.
The width of the violin indicates where fare values are more concentrated.

Question 10 - Insights

1. The Titanic dataset contains more male passengers than female passengers.

2. Most passengers traveled in third class (Pclass 3).

3. Passenger ages are mostly concentrated among young and middle-aged individuals.

4. First-class passengers generally paid higher fares than second- and third-class passengers.

5. Survival was influenced by several factors, including passenger class, fare, age, and gender, indicating that these variables are important in understanding survival patterns.
